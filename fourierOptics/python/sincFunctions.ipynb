```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xVals = np.linspace(-3, 3, 1000)

plt.plot(xVals, np.sinc(xVals)**2, color="indianred")

plt.xlabel("x")
plt.ylabel("sinc$^2$(x)")
```

```python
a = 2
d = 10

def doubleSlit(x):
    return np.sinc(xVals * a / 2)**2 * np.cos(xVals * d)**2

plt.plot(xVals, doubleSlit(xVals), color="indianred")
plt.xlabel("x")
plt.ylabel(r"sinc$^2 \left(\frac{ax}{2} \right) \cos^2(dx)$")
```
