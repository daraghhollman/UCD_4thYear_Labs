```python
import numpy as np
import matplotlib.pyplot as plt

from natsort import natsorted

from matplotlib import cm
from glob import glob

cmap = cm.get_cmap('viridis')
plt.rcParams.update({'font.size': 14})
```

```python
dataPaths = glob("../data/potentiometer/frequencyVariation/*")
dataPaths = natsorted(dataPaths)
print(dataPaths)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(6,6), sharex=True)

# Extract labes from file paths
labels = [f"C={el.split("/")[-1][0:5]}" for el in dataPaths]

for i, path in enumerate(dataPaths):

    data = np.loadtxt(path, skiprows=1)
    
    axes[i].plot(data[:,0]*1000, data[:,1], color=cmap(i/len(axes)), label=labels[i])

    axes[i].set_ylabel("Voltage [V]")
    axes[i].legend(loc="upper right")

plt.xlim(0, 2)
plt.xlabel("Time [ms]")
```
